```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('totals.csv')
alt_df = df.drop(df.tail(1).index) # only so the `plot` does not capture the totals per column
```

```python
df.dtypes
```

### Any trend based on the totals?

```python
alt_df.plot(kind='line', x='period', y='total')
```

### Based on recent events...

I am particularly curious about anti-Asian hate at the onset and height of Covid-19, as well as anti-Jewish sentiments following the Israel-Palestine conflict.

```python
alt_df.plot(x='period', y='anti-asian')
```

```python
alt_df.plot(x='period', y='anti-jewish')
```

### Now, let's try to group them for comparison? Starting with _races_...

```python
alt_df.plot(x='period', y=['anti-asian', 'anti-black', 'anti-arab', 'anti-hispanic', 'anti-white'])
```

```python
plt.figure(figsize=(14, 6)) 

# Plot each column separately
plt.plot(alt_df['period'], alt_df['anti-asian'], label='anti-asian')
plt.plot(alt_df['period'], alt_df['anti-black'], label='anti-black')
plt.plot(alt_df['period'], alt_df['anti-arab'], label='anti-arab')
plt.plot(alt_df['period'], alt_df['anti-hispanic'], label='anti-hispanic')
plt.plot(alt_df['period'], alt_df['anti-white'], label='anti-white')
plt.plot(alt_df['period'], alt_df['anti-other ethnicity'], label='anti-other ethnicity')
plt.plot(alt_df['period'], alt_df['anti-multi racial groups'], label='anti-multi racial groups') 

plt.xlabel('Period')
plt.ylabel('Complaints filed')
plt.title('Anti-racism incidents in New York City, 2020-2024 Q1')

# Add legend
plt.legend()

plt.savefig("racism.svg", format='svg')

plt.show()
```

### What about religion?

```python
plt.figure(figsize=(16, 6)) 

# Plot each column separately
plt.plot(alt_df['period'], alt_df['anti-catholic'], label='anti-catholic')
plt.plot(alt_df['period'], alt_df['anti-hindu'], label='anti-hindu')
plt.plot(alt_df['period'], alt_df['anti-muslim'], label='anti-muslim')
plt.plot(alt_df['period'], alt_df['anti-jewish'], label='anti-jewish')
plt.plot(alt_df['period'], alt_df['anti-religious practice generally'], label='anti-religious practice generally')
plt.plot(alt_df['period'], alt_df['anti-sikh'], label='anti-sikh')
plt.plot(alt_df['period'], alt_df['anti-eastern orthodox'], label='anti-eastern orthodox')
plt.plot(alt_df['period'], alt_df['anti-jehovahs witness'], label='anti-jehovahs witness')
plt.plot(alt_df['period'], alt_df['anti-buddhist'], label='anti-buddhist')
plt.plot(alt_df['period'], alt_df['anti-protestant'], label='anti-protestant')

plt.xlabel('Period')
plt.ylabel('Complaints filed')
plt.title('Anti-religion incidents in New York City, 2020-2024 Q1')

# Add legend
plt.legend()

plt.savefig("religion.svg", format='svg')

plt.show()
```

### Gender identity?

```python
plt.figure(figsize=(16, 6)) 

# Plot each column separately
plt.plot(alt_df['period'], alt_df['anti-lesbian'], label='anti-lesbian')
plt.plot(alt_df['period'], alt_df['anti-gay'], label='anti-gay')
plt.plot(alt_df['period'], alt_df['anti-transgender'], label='anti-transgender')
plt.plot(alt_df['period'], alt_df['anti-female'], label='anti-female')
plt.plot(alt_df['period'], alt_df['anti-gender non-conforming'], label='anti-gender non-conforming')
plt.plot(alt_df['period'], alt_df['anti-lgbt'], label='anti-lgbt')

plt.xlabel('Period')
plt.ylabel('Complaints filed')
plt.title('Anti-LGBT incidents in New York City, 2020-2024 Q1')

# Add legend
plt.legend()

plt.savefig("lgbt.svg", format='svg')

plt.show()
```
